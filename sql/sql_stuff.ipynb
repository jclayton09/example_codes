# Functions

```python
def select_data():
    select = f'''
    SELECT *
    FROM testing.hello_world
    '''
    with create_engine(db).connect() as con:
        df = pd.read_sql(sql=select,
                         con = con,
                        )
    return df

def insert_data(coloumns_values):
    insert = f'''
    INSERT INTO testing.hello_world (id, {coloumns_values[0][0]}, {coloumns_values[1][0]})
    VALUES (DEFAULT, '{coloumns_values[0][1]}'::text, '{coloumns_values[1][1]}'::text);
    '''
    with create_engine(db).connect() as con:
        con.execute(insert)
    return select_data()
```
